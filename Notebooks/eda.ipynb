```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Housing.csv")
print(df.head())
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df['Location_wise_Average_price'] = df.groupby(df['Location'])['Price'].transform('mean')

# print(df['Location_wise_Average_price'])
```

```python
df.drop(['BED','Sofa','TV','DiningTable','Refrigerator','Wardrobe'], axis=1, inplace=True)
```

```python
df.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
df.drop(['Location', 'City'], axis=1, inplace=True)
```

```python
df.drop(['Microwave', 'WashingMachine','GolfCourse'], axis=1, inplace=True)
df.to_csv("Housing.csv")
df.columns
```
